```python
# Import Libraries

import string
```

```python
# Product Database

products = {
    "laptop": "Laptop available for ₹55,000",
    "phone": "Smartphone available for ₹25,000",
    "headphones": "Headphones available for ₹3,000",
    "watch": "Smartwatch available for ₹5,000"
}
```

```python
# Intent Patterns and Responses

intent_patterns = {
    "product_search": [
        "buy", "search", "find", "product",
        "laptop", "phone", "headphones", "watch"
    ],
    
    "order_tracking": [
        "track", "order", "delivery",
        "shipment", "status"
    ],
    
    "return_policy": [
        "return", "refund", "replace",
        "exchange", "policy"
    ]
}

responses = {
    "order_tracking": "Your order is currently in transit and will be delivered soon.",
    
    "return_policy": "Products can be returned within 7 days of delivery subject to return policy terms.",
    
    "unknown": "I'm sorry, I could not understand your request. Please rephrase."
}
```

```python
# Text Preprocessing

def preprocess(text):
    
    text = text.lower()
    
    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )
    
    return text.split()
```

```python
# Intent Detection

def detect_intent(tokens):
    
    scores = {}
    
    for intent, keywords in intent_patterns.items():
        scores[intent] = 0
        
        for word in tokens:
            if word in keywords:
                scores[intent] += 1
    
    best_intent = max(scores, key=scores.get)
    
    if scores[best_intent] == 0:
        return "unknown"
    
    return best_intent
```

```python
# Response Generation

def get_response(user_input):
    
    tokens = preprocess(user_input)
    
    intent = detect_intent(tokens)
    
    if intent == "product_search":
        
        for word in tokens:
            if word in products:
                return products[word]
        
        return "Please specify which product you are looking for."
    
    return responses[intent]
```

```python
# Chatbot Interaction Loop

print("E-Commerce Chatbot Activated")
print("Type 'exit' to stop.\n")

while True:
    
    user_input = input("You: ")
    
    if user_input.lower() == "exit":
        print("Bot: Thank you for visiting. Goodbye!")
        break
    
    response = get_response(user_input)
    
    print("Bot:", response)
```
